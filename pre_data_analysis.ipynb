```python
import os
import shutil
import random
```

```python
data_folder = './data'

for folder_name in os.listdir(data_folder):
    folder_path = os.path.join(data_folder, folder_name)
    
    densepose_folder = os.path.join(folder_path, 'densepose')
    if os.path.exists(densepose_folder):
        shutil.rmtree(densepose_folder)
        print(f'"densepose" folder removed from {folder_name}.')
    else:
        print(f'Folder "densepose" not found in {folder_name}.')
```

```python
total_images = 0

for folder_name in os.listdir(data_folder):
    folder_path = os.path.join(data_folder, folder_name)
    images_folder = os.path.join(folder_path, 'images')
    
    if os.path.exists(images_folder):
        images_count = len(os.listdir(images_folder))
        total_images += images_count

print(f'The total number of pictures in the "images" folders: {total_images}')
```

```python
train_folder = os.path.join(data_folder, 'train')
val_folder = os.path.join(data_folder, 'val')
test_folder = os.path.join(data_folder, 'test')

os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

train_percent = 0.78
val_percent = 0.11
test_percent = 0.11

folder_names = [folder_name for folder_name in os.listdir(data_folder) if os.path.exists(os.path.join(data_folder, folder_name, 'images'))]

random.shuffle(folder_names)

num_folders = len(folder_names)
num_train = int(train_percent * num_folders)
num_val = int(val_percent * num_folders)

train_folders = folder_names[:num_train]
val_folders = folder_names[num_train:num_train + num_val]
test_folders = folder_names[num_train + num_val:]

for folder_name in train_folders + val_folders + test_folders:
    images_folder = os.path.join(data_folder, folder_name, 'images')
    images = os.listdir(images_folder)
    file_prefix = f'{folder_name}_'

    if folder_name in train_folders:
        dst_folder = train_folder
    elif folder_name in val_folders:
        dst_folder = val_folder
    else:
        dst_folder = test_folder

    for file_name in images:
        src_image_path = os.path.join(data_folder, folder_name, 'images', file_name)
        src_mask_path = os.path.join(data_folder, folder_name, 'masks', file_name)

        dst_image_path = os.path.join(dst_folder, 'images', file_prefix + file_name)
        dst_mask_path = os.path.join(dst_folder, 'masks', file_prefix + file_name)

        os.makedirs(os.path.dirname(dst_image_path), exist_ok=True)
        os.makedirs(os.path.dirname(dst_mask_path), exist_ok=True)

        shutil.copyfile(src_image_path, dst_image_path)
        shutil.copyfile(src_mask_path, dst_mask_path)
```
